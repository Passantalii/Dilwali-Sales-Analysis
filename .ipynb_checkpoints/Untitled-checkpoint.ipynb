```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv(r"C:\Users\lap shop\Desktop\Dilwali sales\Diwali Sales Data.csv", encoding = 'unicode_escape')
```

##Cleaning the Data

```python
for col in df.describe(include = object).columns:
    print(col)
    print(df[col].unique())
    print('-'*55)
```

```python
df.info()
```

Removing useless column

```python
df.drop(['Status', 'unnamed1'], axis = 1, inplace = True)
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

Change data types

```python
df['Amount'] = df['Amount'].astype(int)
```

```python
df['Amount'].dtype
```

column rename

```python
df.rename(columns = {'Marital_Status':'Married'}, inplace = True)
```

Data Description (stat)

```python
df[['Age', 'Orders', 'Amount']].describe()
```

EDA

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
data = df.groupby(['Gender'], as_index = False)['Amount'].sum()
```

```python
sns.barplot(x= 'Gender', y = 'Amount', data = data)
```

As shown most purchases done by females (the buyers)

```python
ax = sns.countplot(x = 'Age Group', hue = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

The most age group in males or females is in (26-35) and females are more than males in all age groups

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
total = sum(sales_age['Amount'])
for i, v in enumerate(sales_age['Amount']):
    percent = v/total *100
    plt.text(i, v, f'{percent:.1f}%', ha = "center", va = "bottom")
```

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(11)
```

```python
plt.figure(figsize = (17,7))
sns.barplot(x = 'State', y = 'Orders', data = sales_state)
```

Then we can see where the most of orders and total amount is.

```python
ax = sns.countplot(x = 'Married', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

The unmarried is more than the married buyers

```python
sales_state = df.groupby(['Married', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc = {'figure.figsize':(6,5)})
sns.barplot(x = 'Married', y = 'Amount', hue = 'Gender', data = sales_state)
```

We can see that most of the buyers are not married (women) and they have high purchasing power

```python
sns.set(rc = {'figure.figsize':(20,5)})
ax = sns.countplot(x = 'Occupation', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(x = 'Occupation', y = 'Amount', data = sales_state)
```

We can see that most of the buyers are working in IT sector then Healthcare and Aviation sector.

```python
sns.set(rc = {'figure.figsize':(20,5)})
ax = sns.countplot(x = 'Product_Category', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(11)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(x = 'Product_Category', y = 'Amount', data = sales_state)
```

We can see that most of the sold products are from Food, Clothing and Electronics category

```python
sales_state = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(11)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(x = 'Product_ID', y = 'Orders', data = sales_state)
```

##Conclusion

Unmarried women aged 26-35 from Uttar Pradesh, Maharashtra, and Karnataka working in the IT, Healthcare, and Aviation sectors are more likely to purchase products from the food, clothing, and electronics categories.
